```python
import pandas as pd
import numpy as np
```

```python
review = pd.read_excel('Aish_Updated_Review_Table.xlsx')
review.head()
```

```python
type(review['cusId'][0])
```

```python
write = pd.read_excel('Write.xlsx')
write.head()
```

```python
for i in range(len(review)):
    index_of_revid = write[write['revId'] == review['revId'][i]].index.tolist()
    if len(index_of_revid) == 1:
        review['cusId'][i] = write['cusId'][index_of_revid[0]]
    else:
        continue
        
        
for i in range(len(review)):
    index_of_revid = write[write['revId'] == review['revId'][i]].index.tolist()
    if len(index_of_revid) == 1:
        review['resId'][i] = write['resId'][index_of_revid[0]]
    else:
        continue
  
```
